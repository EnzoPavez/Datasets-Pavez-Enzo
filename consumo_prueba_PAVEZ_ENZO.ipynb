
Este código es útil cuando tengo un archivo de Excel específico y quiero manipular los datos de manera más fácil. Me permite cargar el archivo en Google Colab y usar la biblioteca pandas para trabajar con los datos sin complicaciones.

Además, puedo limpiar el archivo eliminando filas y columnas vacías, lo cual es genial si hay información innecesaria o espacios en blanco en el archivo.

Una vez que hago las modificaciones necesarias, guardo el DataFrame modificado en un nuevo archivo de Excel y lo descargo a mi dispositivo local. A partir de este punto, comienzo a trabajar con el archivo modificado, lo cual me facilita realizar análisis y procesamientos adicionales.

```python
from google.colab import files
import pandas as pd

uploaded = files.upload()

file_name = next(iter(uploaded))
df = pd.read_excel(file_name)

columnas_a_eliminar = ['Unnamed: 1', 'Unnamed: 3', 'Unnamed: 13', 'Unnamed: 23']
df = df.drop(columns=columnas_a_eliminar)

nuevos_nombres = {
    'Unnamed: 6': 'Tiquete',
    'Unnamed: 16': 'Fecha',
    'Unnamed: 18': 'Hora',
    'Unnamed: 19': 'Vehiculo',
    'Unnamed: 20': 'Manifiesto',
    'Unnamed: 21': 'Km',
    'Unnamed: 22': 'Articulo',
    'Unnamed: 25': 'Cant.',
    'Unnamed: 26': 'Precio',
    'Unnamed: 27': 'Total Venta'
}
df = df.rename(columns=nuevos_nombres)
df.insert(0, "Facturas", "")
df["Estacion"] = ""
df = df.reset_index(drop=True)
df["Facturas"] = "COLS"

valores_estacion = {
    168000: "00 SOL",
    80000: "1 SOL",
    416000: "2 SOL",
    184000: "3 SOL",
    30002: "01 AIRES",
    34042: "2 AIRES",
    66912: "3 AIRES",
    553000: "02 URT",
    63000: "3 URT",
    840000: "4 URT",
    350000: "5 URT",
    210000: "6 URT",
    105000: "7 URT"
}
df["Estacion"] = df["Total Venta"].map(valores_estacion)

filas_a_eliminar = [0, 1, 2, 7, 8, 9, 13, 14, 15, 22, 23]
df = df.drop(filas_a_eliminar)

df = df.reset_index(drop=True)

display(df)

new_file_name_csv = 'dataframe_Pavez_Enzo.csv'
df.to_csv(new_file_name_csv, index=False)

files.download(new_file_name_csv)
```
